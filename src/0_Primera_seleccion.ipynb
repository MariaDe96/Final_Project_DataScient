### Creaci√≥n de `tobacco_use`: riesgo acumulativo por uso de nicotina

Se construye una variable `tobacco_use` que unifica el consumo actual de nicotina en diferentes formatos:

- `_RFSMOK3`: fumador actual de cigarrillos tradicionales
- `ECIGNOW2`: uso actual de cigarrillos electr√≥nicos (vapeo)
- `USENOW3`: uso actual de tabaco sin combusti√≥n (snus, masticable, etc.)

#### ‚úÖ Se prefiere este enfoque por:
- Reflejar de forma acumulativa la exposici√≥n **actual** a nicotina, sin limitarse a un solo tipo de producto.
- Capturar mejor el riesgo asociado a salud mental, que est√° m√°s ligado al consumo presente que al hist√≥rico.

#### ‚ùå Se elimina `SMOKE100` porque:
- Mide solo si alguna vez se fumaron 100 cigarrillos (exposici√≥n pasada).
- No distingue entre fumadores actuales y antiguos.
- Su informaci√≥n ya est√° contenida de forma m√°s √∫til en `_RFSMOK3`.

> El resultado es una variable compuesta `tobacco_use` que toma valores de 0 a 3, seg√∫n la cantidad de formas activas de consumo de tabaco o nicotina.

‚ö†Ô∏è **Nota:** antes de realizar la suma, se revisa la codificaci√≥n de cada variable para asegurarse de que los valores est√©n normalizados (por ejemplo, que ‚Äúno consumo‚Äù est√© codificado como 0) y evitar que c√≥digos por defecto (como 7 = ‚Äúno sabe‚Äù o 9 = ‚Äúsin respuesta‚Äù) generen resultados incoherentes.


```python
df_selected = df_selected.drop("SMOKE100", axis=1)
```

```python
df_selected["_RFSMOK3"].value_counts()
```

```python
# Mapeo y recodificacion [No >> 0, Si >> 1, No contesta >> NaN]
df_selected["_RFSMOK3"] = df_selected["_RFSMOK3"].map({
    1:0, 2:1, 9:np.nan
})

df_selected["_RFSMOK3"].value_counts()
```

```python
df_selected["USENOW3"].value_counts()
```

```python
# Mapeo y recodificacion [No >> 0, Si >> 1, No contesta >> NaN]
df_selected["USENOW3"] = df_selected["USENOW3"].map({
    1:1, 2:1, 3:0, 7:np.nan, 9:np.nan
})

df_selected["USENOW3"].value_counts()
```

```python
df_selected["ECIGNOW2"].value_counts()
```

```python
# Mapeo y recodificacion [No >> 0, Si >> 1, No contesta >> NaN]
df_selected["ECIGNOW2"] = df_selected["ECIGNOW2"].map({
    1:0, 2:1, 3:1, 4:0, 7:np.nan, 9:np.nan
})

df_selected["ECIGNOW2"].value_counts()
```

```python
df_selected = df_selected.dropna(subset=["_RFSMOK3", "USENOW3", "ECIGNOW2"])

df_selected["tobacco_use"] = df_selected[["_RFSMOK3", "USENOW3", "ECIGNOW2"]].sum(axis=1)

df_selected = df_selected.drop(["_RFSMOK3", "USENOW3", "ECIGNOW2"], axis=1)
```
