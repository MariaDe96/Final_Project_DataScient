### Creación de `tobacco_use`: riesgo acumulativo por uso de nicotina

Se construye una variable `tobacco_use` que unifica el consumo actual de nicotina en diferentes formatos:

- `_RFSMOK3`: fumador actual de cigarrillos tradicionales
- `ECIGNOW2`: uso actual de cigarrillos electrónicos (vapeo)
- `USENOW3`: uso actual de tabaco sin combustión (snus, masticable, etc.)

#### ✅ Se prefiere este enfoque por:
- Reflejar de forma acumulativa la exposición **actual** a nicotina, sin limitarse a un solo tipo de producto.
- Capturar mejor el riesgo asociado a salud mental, que está más ligado al consumo presente que al histórico.

#### ❌ Se elimina `SMOKE100` porque:
- Mide solo si alguna vez se fumaron 100 cigarrillos (exposición pasada).
- No distingue entre fumadores actuales y antiguos.
- Su información ya está contenida de forma más útil en `_RFSMOK3`.

> El resultado es una variable compuesta `tobacco_use` que toma valores de 0 a 3, según la cantidad de formas activas de consumo de tabaco o nicotina.

⚠️ **Nota:** antes de realizar la suma, se revisa la codificación de cada variable para asegurarse de que los valores estén normalizados (por ejemplo, que “no consumo” esté codificado como 0) y evitar que códigos por defecto (como 7 = “no sabe” o 9 = “sin respuesta”) generen resultados incoherentes.


```python
df_selected = df_selected.drop("SMOKE100", axis=1)
```

```python
df_selected["_RFSMOK3"].value_counts()
```

```python
# Mapeo y recodificacion [No >> 0, Si >> 1, No contesta >> NaN]
df_selected["_RFSMOK3"] = df_selected["_RFSMOK3"].map({
    1:0, 2:1, 9:np.nan
})

df_selected["_RFSMOK3"].value_counts()
```

```python
df_selected["USENOW3"].value_counts()
```

```python
# Mapeo y recodificacion [No >> 0, Si >> 1, No contesta >> NaN]
df_selected["USENOW3"] = df_selected["USENOW3"].map({
    1:1, 2:1, 3:0, 7:np.nan, 9:np.nan
})

df_selected["USENOW3"].value_counts()
```

```python
df_selected["ECIGNOW2"].value_counts()
```

```python
# Mapeo y recodificacion [No >> 0, Si >> 1, No contesta >> NaN]
df_selected["ECIGNOW2"] = df_selected["ECIGNOW2"].map({
    1:0, 2:1, 3:1, 4:0, 7:np.nan, 9:np.nan
})

df_selected["ECIGNOW2"].value_counts()
```

```python
df_selected = df_selected.dropna(subset=["_RFSMOK3", "USENOW3", "ECIGNOW2"])

df_selected["tobacco_use"] = df_selected[["_RFSMOK3", "USENOW3", "ECIGNOW2"]].sum(axis=1)

df_selected = df_selected.drop(["_RFSMOK3", "USENOW3", "ECIGNOW2"], axis=1)
```
